```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.pipeline import Pipeline

from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('/content/store_customers.csv')
df.sample(5)
```

```python
df.isnull().sum()
```

```python
median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)
```

```python
median_annual_income = df['Annual Income (k$)'].median()
df['Annual Income (k$)'] = df['Annual Income (k$)'].fillna(median_annual_income)
```

```python
median_spending_score = df['Spending Score (1-100)'].median()
df['Spending Score (1-100)'] = df['Spending Score (1-100)'].fillna(median_spending_score)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
x = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
# split columns for scaling and encoding
nums = ['Annual Income (k$)','Spending Score (1-100)']
```

```python
preprocessor = ColumnTransformer([
    ("nums",StandardScaler(),nums)
])
```

```python
#KMeans pipeline
pipeline = Pipeline([
    ('preprocess',preprocessor),
    ('model',KMeans(n_clusters=5,random_state=42))
])
```

```python
label_kmeans = pipeline.fit_predict(x)
```

```python
#DBSCAN pipeline
pipeline_dbscan = Pipeline([
    ("preprocess",preprocessor),
    ("model",DBSCAN(eps=0.2,min_samples=2))
])
```

```python
label_dbscan = pipeline_dbscan.fit_predict(x)
```

```python
def evlauation(x,labels,name):
  if len(set(labels)) > 1:
    score = silhouette_score(preprocessor.fit_transform(x),labels)
    print(f"{name} silhoutte score is:",score)
  else:
    print(f"{name} failed (single cluster)")
```

```python
evlauation(x,label_kmeans,"Kmeans")
evlauation(x,label_dbscan,"DBSCAN")
```
